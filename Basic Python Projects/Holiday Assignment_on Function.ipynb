# 3. Write a function which will take input as dict. and give me output a list of all        the values of even in case of two level nesting should work. 

```python
def test3(a):
    '''taking input as dict. values and giving output as list values'''
    c=[]
    for i in a.values():
        if type(i) == int:
            if i % 2 == 0:
                c.append(i)
        elif type(i)==str or type(i)==set or type(i)==tuple:
            c.append(i)
    print(c)
```

# 10. Write a function which will read video file and play for you.

```python
import cv2

def task11_video(video_path):
    video = cv2.VideoCapture(video_path)
    
    while(video.isOpened()):
      # Capturing the video frame-by-frame
        ret, frame = video.read()
        if ret == True:
            cv2.imshow('Frame',frame) #Showing the display of current frame
            if cv2.waitKey(25) & 0xFF == ord('x'):  ## Press x for exit the video
                break
        else:
            break
    video.release()
    cv2.destroyAllWindows()
```

```python
task11_video('video.mp4')
```

# 14. Write a function by which i can send you mail to anyone.

```python
import smtplib
```

```python
def task15_writeMail(user, passwd):
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()  ## creating a connection between gmail and python using tls(transport layer security)
    server.login(user,passwd)  ## Login into the gmail account
    rx_mail = input("Enter receiver mail id: ")
    msg = input('Enter the mail(body): ')
    server.sendmail(user, rx_mail, msg)
    server.close()
    print('Mail Sent')
```

```python
user = 'barishsawant7@gmail.com'
passwd = ''

task15_writeMail(user, passwd)
```

# 15. Write a function to read out complete PDF file.

```python
!pip install PyPDF2
```

```python
import PyPDF2

def task16_readPDF(pdf_path):
    pdfFileObj = open(pdf_path, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    print("Number of pages: ", pdfReader.numPages)
    
    for i in range(0,pdfReader.numPages):
        print("\nPageNo. ", i)
        pageObj = pdfReader.getPage(i)
        print(pageObj.extractText())
        
    pdfFileObj.close()
```

```python
task16_readPDF("mysql-tutorial.pdf")
```

# 19. Write a function by which you will able to append two PDF files.

```python
import PyPDF2 

def task20_mergePDF(file1, file2):
    mergeFile = PyPDF2.PdfFileMerger()
    mergeFile.append(PyPDF2.PdfFileReader(file1, 'rb'))
    mergeFile.append(PyPDF2.PdfFileReader(file2, 'rb'))
    mergeFile.write("NewPDF.pdf")
    print('Merge done.')
```

```python
file1 = 'mysql-tutorial.pdf'
file2 = 'sample_1.pdf'
task20_mergePDF(file1, file2)
```
