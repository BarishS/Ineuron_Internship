# 10. from a cabin column seperate text and numeric value

### Sol:

```python
df.head()
```

```python
df['Cabin']
```

```python
char = []
num = []

for i in df['Cabin']:
    string1 = ' '
    string2 = ' '
    if type(i) == str:
        for ch in i:
            if ch >= 'A' and ch<= 'Z' or ch >= 'a' and ch <= 'z' :
                string1 += ch
            
            else : 
                string2 += ch
        
        char.append(string1)
        num.append(string2)
        
    else:
        char.append(i)
        num.append(i)
        
```

```python
char[: 5]
```

```python
num[:5]
```

```python
df['Cabin_char'] = char
df['Cabin_num'] = num
```
