# Dump all the data into car table 

```python
## Data Preview
with open('car.data', 'r') as f:
    car_data = csv.reader(f, delimiter = '\n')
    for i in car_data:
        row_data = i[0].split(',')
        print(row_data)
```

```python
with open('car.data', 'r') as f:
    car_data = csv.reader(f, delimiter = '\n')
    for i in car_data:
        row_data = tuple(i[0].split(','))
        cursor.execute("INSERT INTO car_db.car VALUES {}".format(row_data))
mydb.commit()
```
